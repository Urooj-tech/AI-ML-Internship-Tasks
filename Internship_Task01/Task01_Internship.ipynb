```python
#  Step# 01. Import Required Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Enable better visuals
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Step# 02. Load the Iris Dataset
# Loading dataset using seaborn's built-in dataset
iris = sns.load_dataset('iris')
```

```python
# Step# 03. Dataset Inspection
print("\nDataset Shape:", iris.shape)
print("\nColumn Names:", iris.columns.tolist())
print("\nFirst 5 Rows:\n", iris.head())

print("\nDataset Info:")
iris.info()
print("\nSummary Statistics:\n", iris.describe())
```

```python
# Step04. Data Visualization

# --- 4.1 Scatter Plot Matrix (Pairplot) ---
sns.pairplot(iris, hue='species', corner=True, palette='Set1')
plt.suptitle("Relationships between Features - Pairplot", y=1.02)
plt.show()
```

```python
# --- 4.2 Histograms ---
iris.hist(edgecolor='black', linewidth=1.2, figsize=(10, 8))
plt.suptitle("Feature Distributions - Histograms", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# --- 4.3 Box Plots ---
features = iris.columns[:-1]
plt.figure(figsize=(14, 6))
for idx, feature in enumerate(features):
    plt.subplot(1, 4, idx+1)
    sns.boxplot(y=iris[feature], color='lightgreen')
    plt.title(f'Boxplot: {feature}')
plt.tight_layout()
plt.show()
```

#### **Skills Applied:**  
- Data loading and inspection using pandas  
- Descriptive statistics and data exploration   
- Basic plotting and visualization with seaborn and matplotlib   

  > ## Summary And Insights
- Dataset contains **150** instances, evenly divided across **3** species.
- Features are numeric, well-behaved (no missing values).
- Petal features show strong class separation.
- Boxplots help identify that petal width has the most variance.
- Scatter plots reveal petal_length and petal_width are useful for classification.
